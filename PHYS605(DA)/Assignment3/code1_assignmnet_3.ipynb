```python
# Importing useful functions
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the triangular distribution function
def triangular_distribution(x, alpha, beta):
    if alpha <= x <= beta:
        return 2 * (x - alpha) / (beta - alpha)**2
    else:
        return 0
def mean_median(alpha,beta):
    mean = (2 * beta + alpha) / 3
    median = alpha + (beta - alpha) / np.sqrt(2)
    return mean, median
# Calculate mode
def calculate_mode(alpha, beta):
    x = np.linspace(alpha, beta,1000)
    y = [triangular_distribution(xi, alpha, beta) for xi in x]
    mode_index = np.argmax(y)
    mode_value = x[mode_index]
    return mode_value

# Plot the triangular distribution
def plot_triangular_distribution(alpha, beta):
    x = np.linspace(alpha-0.5, beta+0.5, 1000)
    y = [triangular_distribution(xi, alpha, beta) for xi in x]

    global mean, median, mode
    mean, median = mean_median(alpha, beta)
    mode = calculate_mode(alpha, beta)

    plt.plot(x, y,  label='Triangular Distribution')
    plt.axvline(mean, color='r', linestyle='-', label='Mean')
    plt.axvline(median, color='g', linestyle='--', label='Median')
    plt.axvline(mode, color='b', linestyle='--', label='Mode')
    plt.legend()
    plt.xlabel('x')
    plt.ylabel('p_T(x)')
    plt.title('Triangular Distribution with Mean, Median, and Mode')
    plt.show()

# Set values for alpha and beta
alpha = 0
beta = 1
# Print moments, median, and mode
print(f"Mean: {mean:.2f}")
print(f"Median: {median:.2f}")
print(f"Mode: {mode:.2f}")

# Plot the distribution
plot_triangular_distribution(alpha, beta)
```
